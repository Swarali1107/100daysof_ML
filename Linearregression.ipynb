```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], color='blue', marker='o')
plt.xlabel('years')
plt.ylabel('per capita income(US$)')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict(pd.DataFrame({'year': [2005]}))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
828.46507522*2005+-1632210.7578554575
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], color='blue', marker='o')
plt.xlabel('years')
plt.ylabel('per capita income(US$)')
plt.plot(df['year'], reg.predict(df[['year']]), color='red', label='Regression Line')
```
